```python
import os
from datasets import load_dataset

# 定义scratch目录路径（请替换'yourusername'为实际用户名）
scratch_path = '/scratch/zdiao7/OLIVES_Dataset'

# 确保目录存在
os.makedirs(scratch_path, exist_ok=True)

# 下载数据集到指定的scratch目录
olives = load_dataset(
    'gOLIVES/OLIVES_Dataset', 
    'disease_classification', 
    split='train',
    cache_dir=scratch_path
)

# 验证下载
print(f"数据集已下载到: {scratch_path}")
print(f"数据集样本数: {len(olives)}")
```

```python
# 转换数据集为PyTorch可用格式
olives = olives.with_format("torch")
```

```python
from torch.utils.data import DataLoader
# 创建DataLoader，批次大小为4
dataloader = DataLoader(olives, batch_size=4)

# 遍历数据集批次
for batch in dataloader:
    print(batch)  # 打印每个批次的内容
```
